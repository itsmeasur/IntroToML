```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/home/aakash/Desktop/MR data - Compiled Data Set.csv')
data
```

```python
data = data.fillna(0)
data = data.drop(['Title','Text'], axis = 1)
old_col = data.columns
data.columns = [i for i in range(0,len(data.columns))]
data
```

```python
data[11] = 0
data[12] = 0

for i in range(0,len(data)):
    if data.iloc[i,8] != 0 : 
        x = str(data.iloc[i,8]).split(',')
        data.iloc[i,8] = int(x[0])
        if len(x) > 1:
            data.iloc[i,11] = int(x[1])
        if len(x) > 2:    
            data.iloc[i,12] = int(x[2])
data
```

```python
data = data[[0,1,2,3,4,5,6,7,8,11,12,9,10]]
data.columns = [i for i in range(0,len(data.columns))]
data
```

```python
total_crimes_citywise = data.groupby(0).size().reset_index(name = 1)
total_crimes_citywise
```

```python
x = [0,0.15,0]
plt.pie(total_crimes_citywise[1], explode = x, labels = total_crimes_citywise[0], shadow = True)
plt.axis('equal')
plt.legend(title = "Cities", bbox_to_anchor=(0.9,1), loc="upper left")
plt.show()
```

```python
s = data[1].str.contains("LUCKNOW", na = False)
for i in range(0, len(data)):
    if s[i] == True:
        data[1] = data[1].replace(data.iloc[i,1], 0)
        
data.iloc[315,1] = data.iloc[315,2]
data.iloc[315,2] = 0
data[2] = data[2].astype(int)
        
reasons_for_murder = data.groupby(1).size().reset_index(name = 2)
reasons_for_murder
```

```python
reasons_for_murder = reasons_for_murder.drop([0])
reasons_for_murder
```

```python
x = [0,0,0.15,0,0,0,0,0]
plt.pie(reasons_for_murder[2], explode = x, labels = reasons_for_murder[1], shadow = True, startangle = 85)
plt.axis('equal')
plt.legend(title = "Reasons for murder", bbox_to_anchor=(1.1,1), loc="upper left")
plt.show()
```

```python
d = data.groupby(1).sum()
d = d.drop([0])
d[2] = d[2]+d[3]+d[4]
reason_for_murder_most_affecting_the_society = [x*y for x,y in zip(d[2],reasons_for_murder[2])]
reason_for_murder_most_affecting_the_society
```

```python
fig = plt.figure(figsize = (10,5))
plt.bar(reasons_for_murder[1],reason_for_murder_most_affecting_the_society)
plt.xlabel("Reasons for murder")
plt.ylabel("Murders committed per reason * Number of people murdered")
plt.xticks(rotation = 45)
plt.title("Reason for murder most affecting the society")
plt.show()
```

```python
pairs = []
for i in range(len(data)):
    if (data[8][i] and data[9][i] and data[10][i]):
        pairs.append([data[8][i],data[9][i],data[10][i]])
        pairs.append([data[8][i],data[9][i]])
        pairs.append([data[9][i],data[10][i]])
        pairs.append([data[8][i],data[10][i]])
    elif (data[8][i] and data[9][i]):
        pairs.append([data[8][i],data[9][i]])
    elif (data[9][i] and data[10][i]):
        pairs.append([data[9][i],data[10][i]])
    elif (data[8][i] and data[10][i]):
        pairs.append([data[8][i],data[10][i]])
pairs = pd.Series(pairs)
pairs
```

```python
x = pd.DataFrame(pairs.value_counts().reset_index(name = 'freq'))
x
```

```python
x['index'] = x['index'].astype(str)
```

```python
fig = plt.figure(figsize = (25,15))
plt.bar(x['index'],x['freq'])
plt.xlabel("Crimes happening simultaneously")
plt.ylabel("Frequency")
plt.title("Crimes against women")
plt.show()
```
